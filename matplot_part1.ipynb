```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from IPython.display import Image
%matplotlib inline
mpl.__version__
```

```python
print(plt.style.available)
```

```python
def criar_plot():
    x = np.random.randn(5000,6)
    (figure, axes) = plt.subplots(figsize = (10,4))
    (n, bins, patches) = axes.hist(x, 8,
                                   density = 1,
                                   histtype = 'bar',
                                   label = ['Color 1', 'Color 2', 'Color 3', 'Color 4', 'Color 5', 'Color 6'])
    axes.set_title("Histogram\nPara\nDistribuição Normal", fontsize = 16)
    axes.set_xlabel("Dados", fontsize = 16)
    axes.set_ylabel("Frequência", fontsize = 16)
    axes.legend()
    plt.show()
```

```python
criar_plot()
```

```python
def gera_plot():
    x = np.random.randn(5000, 6)
    (figure, axes) = plt.subplots(figsize= (10, 4))
    (n, bins, patches) = axes.hist(x, 8,
                                  density = 1,
                                  histtype = 'bar',
                                  label = ["Color 1", "Color 2", "Color 3","Color 4","Color 5","Color 6"])
    axes.set_title("Histograma", fontsize=16)
    axes.set_xlabel("Dados", fontsize=16)
    axes.set_ylabel("Frequencia", fontsize=16)
    axes.legend()
    plt.show()
```

```python
gera_plot()
```
